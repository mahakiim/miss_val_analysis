```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('D:\Paper_Imputation\data\dataset_mitigasi_banjir_fix_banget.csv')

```

```python
data.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

df_mice = data.copy()
```

```python
df_mice.head()
```

```python
numeric_cols = ['curah_hujan_DK', 'curah_hujan_BS', 'debit_sungai_m3_per_detik', 'tinggi_muka_air_m']
mice_imputer = IterativeImputer(random_state=123)
df_mice[numeric_cols] = mice_imputer.fit_transform(df_mice[numeric_cols])
```

```python
print("Jumlah missing values setelah MICE/Iterative Imputer dengan random_state=123:")
print(df_mice.isnull().sum())
```
